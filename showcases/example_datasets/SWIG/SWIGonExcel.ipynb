# transforming SWIG

```python
import frictionless
from frictionless import steps
```

```python
SWIG = frictionless.Package("../catalogue/temp_3/SWIG-database/SWIG database.xlsx")
#SWIG
```

# Resource Metadata

```python
SWIG.resources[4].dialect.comment_rows = [1]
SWIG.resources[4].schema = {}
SWIG.resources[4].infer()
for x in ['Gravel','FC','PWP','WC_r','MWD','GMD','pH','Gypsum','CCE','CEC', 'SAR', 'DiscRedius', 'Vegetation cover', 'Rainfal intensity', 'Slope', 'field39']:
    SWIG.resources[4].schema.update_field(name=x, descriptor={'type': 'number'})
for x in ['Treatment', 'Crust', 'Sand contact layer']:
    SWIG.resources[4].schema.update_field(name=x, descriptor={'type': 'string'})
SWIG.resources[4].schema.missing_values +=" "
frictionless.transform(SWIG.resources[4], steps = [steps.field_remove(names = 'field39')])

SWIG.resources[4]
```
