# transforming SWIG

```python
import frictionless
SWIG = frictionless.Package()
```

```python
# import resources
detector = frictionless.Detector(buffer_size=500000)
meta = frictionless.describe(
    "../catalogue/temp_3/SWIG-database/Metadata.csv",
    dialect = frictionless.Dialect(comment_rows = [1]),
    detector = detector)
meta.infer()
```

```python
# at some cells a blank space is used as missing value...
meta.schema.missing_values +=" "
meta.validate()
```

```python
meta.validate()
meta.extract()
```

```python
from frictionless import steps
frictionless.transform(meta, steps = [steps.field_remove(names = 'field39')])
```
